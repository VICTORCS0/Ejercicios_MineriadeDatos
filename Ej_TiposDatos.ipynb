```python
def factorial(n):
    resultado = 1

    for i in range(1, n + 1):
        resultado *= i

    return resultado

num = int(input("Ingresa un número: "))
print("Factorial:", factorial(num))
```

```python
def calcular_factura(cantidad, iva=21):
    total = cantidad + (cantidad * iva / 100)
    return total

cantidad = float(input("Ingresa la cantidad sin IVA: "))

print("Total con IVA:", calcular_factura(cantidad))
```

```python
import math

def area_circulo(radio):
    return math.pi * radio ** 2

def volumen_cilindro(radio, altura):
    return area_circulo(radio) * altura

radio = float(input("Ingresa el radio: "))
altura = float(input("Ingresa la altura: "))

print("Área del círculo:", area_circulo(radio))
print("Volumen del cilindro:", volumen_cilindro(radio, altura))
```

```python
import csv

total = 0

with open("ventas.csv","r") as archivo:
  lector = csv.DictReader(archivo)

  for fila in lector:
    #print(fila)
    cantidad = int(fila["Cantidad"])
    precio = float(fila["Precio_Unitario"])
    total += cantidad * precio

print("Ventas totales:",total)
```

```python
import csv

archivo = open("calificaciones.csv", "r")
lector = csv.reader(archivo)

next(lector)  # Saltar encabezados

suma_promedios = 0
cantidad_alumnos = 0

for fila in lector:
    nombre = fila[1]

    parcial1 = float(fila[3])
    parcial2 = float(fila[4])
    parcial3 = float(fila[5])

    promedio = (parcial1 + parcial2 + parcial3) / 3

    print(nombre, "Promedio:", promedio)

    suma_promedios = suma_promedios + promedio
    cantidad_alumnos = cantidad_alumnos + 1

promedio_general = suma_promedios / cantidad_alumnos

print("Promedio general:", promedio_general)

archivo.close()
```

```python
import csv

archivo = open("calificaciones.csv", "r")
lector = csv.reader(archivo)

next(lector)

mejor_nombre = ""
mejor_promedio = 0

peor_nombre = ""
peor_promedio = 100

for fila in lector:
    nombre = fila[1]

    p1 = float(fila[3])
    p2 = float(fila[4])
    p3 = float(fila[5])

    promedio = (p1 + p2 + p3) / 3

    if promedio > mejor_promedio:
        mejor_promedio = promedio
        mejor_nombre = nombre

    if promedio < peor_promedio:
        peor_promedio = promedio
        peor_nombre = nombre

print("Mejor alumno:", mejor_nombre, mejor_promedio)
print("Peor alumno:", peor_nombre, peor_promedio)

archivo.close()
```

```python
import csv

archivo = open("calificaciones.csv", "r")
lector = csv.reader(archivo)

next(lector)

aprobados = 0
reprobados = 0

for fila in lector:

    p1 = float(fila[3])
    p2 = float(fila[4])
    p3 = float(fila[5])

    promedio = (p1 + p2 + p3) / 3

    if promedio >= 70:
        aprobados = aprobados + 1
    else:
        reprobados = reprobados + 1

print("Aprobados:", aprobados)
print("Reprobados:", reprobados)

archivo.close()
```
